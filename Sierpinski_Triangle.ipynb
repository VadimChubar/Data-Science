```python
import matplotlib.pyplot as plt
from random import randint

#
x1 = 0
y1 = 500
x2 = 500
y2 = 500
x3 = 250
y3 = 0

#
p1 = 250
p2 = 335

#
new_p1 = 0
new_p2 = 0

# +     
list_coord_triangle = [(0, 500), (500, 500), (250, 0)]

triangle = [plt.scatter(x1, y1, marker = '.', color = 'green'), #
            plt.scatter(x2, y2, marker = '.', color = 'green'), #
            plt.scatter(x3, y3, marker = '.', color = 'green'), #
            plt.scatter(p1, p2, marker = '.', color = 'green') #
           ]


i = 0 

while i <= 10000:
    
    #
    random_coord_triangle = list_coord_triangle[randint(0, 2)]
    x = random_coord_triangle[0]
    y = random_coord_triangle[1]
    
    #
    new_p1 = (p1 + x) / 2 
    new_p2 = (p2 + y) / 2 
    
    triangle.append(plt.scatter(new_p1, new_p2, marker = '.', color = 'green'))
    
    p1 = new_p1
    p2 = new_p2
    
    i = i +1
    
plt.show()
```
