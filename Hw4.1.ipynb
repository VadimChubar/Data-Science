```python
# Є набір даних speed, який представляє значення швидкості для деякого транспортного засобу в певний момент спостереження. 
# Очевидно, що дані мають дискретний вигляд. 
# Відомо, що спостереження відбувалися з періодом в одну годину.
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
from scipy import integrate
```

```python
speed = [25, 35, 45, 30, 60, 120, 100, 100, 70, 75, 80, 65]
```

```python
# Приймемо, що в нульовій координаті у нас швидкість 25 км/год, відповідно до набору даних speed. 
# Помістіть у змінну time - вектор часу, створений за допомогою np.linspace (всього 12 спостережень, від 0 до 11 годин)
```

```python
time = np.linspace(0, 11, num=12)
```

```python
# Виконайте виведення масиву time
print(time)
```

```python
# Виконайте виведення графіка — значення точок швидкості (plot або scatter). 
# Задайте розмір області, що відображається (0, 11) і (0, 130). 
# Задайте відображення сітки

plt.plot(time, speed)
plt.ylim(0, 130) # межі для осі y
plt.xlim(0, 11) # межі для осі x
plt.title("Значення точок швидкості", fontsize=10)
plt.xlabel("Швидкість", fontsize="small")
plt.ylabel("Час", fontsize="small")
plt.grid() #сітка
plt.show()
```

```python
# Виконайте інтерполяцію за допомогою interp1d(kind='cubic') та отримайте функцію на 10000 значень. 
# Побудуйте безперервний графік отриманої функції.


time = np.linspace(0, 11, num=12)
sppeed = [25, 35, 45, 30, 60, 120, 100, 100, 70, 75, 80, 65]

f = interp1d(time, speed, kind='cubic')

plt.plot(time, speed, 'o')
more_date = np.linspace(0, 11, 1000)

plt.plot(more_date, f(more_date))

plt.grid() 
plt.show()
```

```python
# Виконайте також пункти 4 та 5 для kind='quadratic'

time = np.linspace(0, 11, num=12)
sppeed = [25, 35, 45, 30, 60, 120, 100, 100, 70, 75, 80, 65]

f = interp1d(time, speed, kind='quadratic')

plt.plot(time, speed, 'o')
more_date = np.linspace(0, 11, 1000)

plt.plot(more_date, f(more_date))

plt.grid() 
plt.show()
```

```python
# Обчисліть інтеграл для отриманої інтерполяційної функції на проміжку [0, 11]

time = np.linspace(0, 11, num=12)
sppeed = [25, 35, 45, 30, 60, 120, 100, 100, 70, 75, 80, 65]

f = interp1d(time, speed, kind='quadratic')

result = integrate.quad(lambda f: f, 0, 11)
print(result)
```
